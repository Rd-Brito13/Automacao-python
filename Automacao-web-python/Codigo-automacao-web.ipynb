```python
#passo a passo
import pandas as pd
from selenium import webdriver as wd

#Passo 1: Abrir navegador
navegador = wd.Chrome()
navegador.get("https://www.google.com.br/")
```

```python
#Passo 2:Importar base de dados
tabela = pd.read_excel("commodities.xlsx")
```

```python
#Passo 3:pegar preço atual do produto, preencher o preço atual na base de dados.

#estrutura de repitção para capturar os valores de cada produto na ordem sequencial da coluna produtos
for linhas in tabela.index:
    produto = tabela.loc[linhas, "Produto"]
    #nesta etapa, exibimos o link e tiramos os caracteres epeciais para quando ocorrer a busca pelo link na web.
    link = (f"https://www.melhorcambio.com/{produto}-hoje")
    link = link.replace("ó", "o").replace("ã", "a").replace("ç", "c").replace("ú", "u").replace("é","e").replace("á","a")
    print(link)
    navegador.get(link)
    
    #buscar a cotação do produto e formatar o valor retornado
    cotacao = navegador.find_element("xpath", '//*[@id="comercial"]').get_attribute("value")
    #retirando a virgula e substituindo por espaço vazio e adicionando . ao valor
    cotacao = cotacao.replace(".", "").replace(",", ".")
    #formatar o valor retornado em float
    cotacao = float(cotacao)
    #printando o valor da cotação retornado e formatado
    print(cotacao)
    #função utilizada para adicionar o valor da cotacao retornada do site a base de dados.
    tabela.loc[linhas,"Preço Atual"] = cotacao
#printar a base de dados atualizada
display(tabela)
   
```

```python
#Passo 4:decidir quais produtos iremos comprar
tabela["Comprar"] = tabela["Preço Atual"] < tabela["Preço Ideal"]
display(tabela)
```

```python
navegador.quit()
```

```python
#Passo 5:exportar a base de dados atualizada
tabela.to_excel("Automacao-web.xlsx", index=False)
```
