# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
import pyperclip


#pyautogui.click > clicar com o mouse
#pyautogui.write > digitar um texto, link.
#pyautogui.hotkey > utilizar teclas de atalho no teclado (ex: ctrl+v, ctrl+c...)
#pyautogui.press > pressionar uma tecla no teclado

pyautogui.PAUSE = 3.5
#Passo - 1: Acessar o sistema da empresa.
pyautogui.hotkey("win")
pyautogui.write("Brave")
pyautogui.press("enter")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

time.sleep(5)

#Passo - 2:Fazer login no sistema.
#clicar e digitar o loguin.
pyautogui.click(x=888, y=336)
pyautogui.write("Meu_Loguin")
#clicar e digitar a senha.
pyautogui.click(x=874, y=426)
pyautogui.write("Minha_Senha")
#clicar no botão para confirmar o acesso.
pyautogui.click(x=928, y=493)

#clicar no arquivo desejado.
pyautogui.click(x=305, y=306)
#iniciar o download.
pyautogui.click(x=507, y=161)
#escolher local de dowload e confirmar.
time.sleep(3)
pyautogui.click(x=1149, y=898)





```

```python
#Passo - 3: Calcular indicadores.
import pandas


tabela = pandas.read_csv(r"C:\Users\RD\Downloads\Compras.csv", sep=";")
display(tabela)

Total_Gasto = tabela["ValorFinal"].sum()
Quantidade = tabela["Quantidade"].sum()
Valor_Media= Total_Gasto / Quantidade

print(Total_Gasto)
print(Quantidade)
print(Valor_Media)
```

```python
#Passo - 4: enviar e-mail para diretoria/chefe.
pyautogui.hotkey("ctrl","t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")
time.sleep(5)

pyautogui.click(x=101, y=163)
time.sleep(5)

pyautogui.click(x=1333, y=521)
pyautogui.write("charmander1759@gmail.com")
pyautogui.press("tab")
pyautogui.click(x=1350, y=583)
pyautogui.write("Relatório de compras")
pyautogui.click(x=1310, y=606)
texto = f"""
Prezados,
Segue o relatório de compras.
Total Gasto: {Total_Gasto:,.2f}
Quantidade Total: {Quantidade:,.2f}
Preço médio: {Valor_Media:,.2f}

Qualquer dúvida, entrar em contato.

Att,.

Rodrigo.
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")

time.sleep(2)
pyautogui.click(x=1417, y=1042)
pyautogui.click(x=958, y=867)
pyautogui.write(r"C:\Users\RD\Downloads\Compras.csv")
pyautogui.click(x=1120, y=894)
pyautogui.hotkey("ctrl", "enter")








```
